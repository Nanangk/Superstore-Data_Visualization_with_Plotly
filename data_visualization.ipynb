# Scatter Plot

```python
data = go.Scatter(
    x=df.sales,
    y=df.profit,
    mode="markers",
    name="Sales VS Profit"
)

fig = go.Figure(data=data)
iplot(fig)
```

```python
df.head()
```

add new coloum

```python
df["marker_color"] = np.where(
    df.profit >= 0,
    "green",
    "red"
)

df.head()

```

Scatter plot color

```python
data = go.Scatter(
    x=df.sales,
    y=df.profit,
    marker={
        "color" : df.marker_color
    },
    mode="markers",
    name="Sales VS Profit"
)

layout={
    "title":{
        "text":"Sales VS Profit",
        "x" : 0.5
    },
    
     "xaxis" : {
        "title" : "Sales",
         
    },
    
    "yaxis" : {
        "title" : "Profit",
       
    },
}



fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# Piechart

```python
data = go.Pie(labels=df.segment, values=df.sales)
layout={
    "title":{
        "text":"Segment Sales ",
        "x" : 0.45
    },
}

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# Histogram 

```python
data = go.Histogram(
    x=df.sales,
    xbins={
        "size" : 500,
    }
)

fig = go.Figure(data=data)
iplot(fig)
```

# Bar Chart

```python
df_segment = df.groupby("segment").agg({"sales" : "sum"})
df_segment = df_segment.reset_index()
df_segment
```

```python
data = go.Bar(
    x = df_segment.segment,
    y = df_segment.sales
)

layout={
    "title":{
        "text":"Segment Sales",
        "x" : 0.5
    },
    
     "xaxis" : {
        "title" : "Segment",
         
    },
    
    "yaxis" : {
        "title" : "Sales",
       
    },
}

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
dfsc = df.groupby(["segment", "category"]).agg({"sales" : "sum"})
dfsc = dfsc.reset_index()
dfsc
```

# Group Bar Chart

```python
bar1 = go.Bar(
    x = dfsc[dfsc.segment == "Corporate"].category,
    y = dfsc[dfsc.segment == "Corporate"].sales,
    name = "Corporate"
)

bar2 = go.Bar(
    x = dfsc[dfsc.segment == "Consumer"].category,
    y = dfsc[dfsc.segment == "Consumer"].sales,
    name = "Consumer"
)

bar3 = go.Bar(
    x = dfsc[dfsc.segment == "Home Office"].category,
    y = dfsc[dfsc.segment == "Home Office"].sales,
    name = "Home Office"
)

layout={
    "title":{
        "text":"Sales Category",
        "x" : 0.5
    },
    
     "xaxis" : {
        "title" : "Category",
         
    },
    
    "yaxis" : {
        "title" : "Sales",
       
    },
}

data = [bar1, bar2, bar3]
fig = go.Figure(data=data)
iplot(fig)
```

# Line Chart

```python
#add month coloum
df["month"] = df.order_date.dt.strftime("%Y-%m-01")
df.head()
```

```python
#groupby sales/month
df_montly_sale = df.groupby("month").agg({"sales" : "sum"})
```

```python
df_montly_sale = df_montly_sale.reset_index()
df_montly_sale.head()
```

```python
data = go.Scatter(
    x = df_montly_sale.month,
    y = df_montly_sale.sales,
    mode = "lines+markers"
)

layout={
    "title":{
        "text":"Monthly Sales",
        "x" : 0.5
    },
    
     "xaxis" : {
        "title" : "Month",
         
    },
    
    "yaxis" : {
        "title" : "Sales",
       
    },
}
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df_monthly_sales_category = df.groupby(["month", "category"]).agg({"sales" : "sum"})
df_monthly_sales_category = df_monthly_sales_category.reset_index()
df_monthly_sales_category.head()
```

```python
df_monthly_sales_category = df_monthly_sales_category.pivot(
    index = "month",
    columns = "category",
    values = "sales"
)

df_monthly_sales_category.head()
```

Multiple Line Chart

```python
line1 = go.Scatter(
    x = df_monthly_sales_category.index,
    y = df_monthly_sales_category["Furniture"],
    mode = "lines+markers",
    marker = {"symbol" : "diamond"},
    name = "Furniture"
)

line2 = go.Scatter(
    x = df_monthly_sales_category.index,
    y = df_monthly_sales_category["Office Supplies"],
    mode = "lines+markers",
     marker = {"symbol" : "cross"},
    name = "Office Supplies"
)

line3 = go.Scatter(
    x = df_monthly_sales_category.index,
    y = df_monthly_sales_category["Technology"],
    mode = "lines+markers",
    name = "Technology"
)

layout={
    "title":{
        "text":"Monthly Sales Category",
        "x" : 0.5
    },
    
    "xaxis" : {
        "title" : "Month",
         
    },
    
    "yaxis" : {
        "title" : "Sales",
       
    },
   
}


data = [line1, line2, line3]
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# Trend Line

```python
from scipy import stats
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x=df.sales, y=df.profit)
lines = slope * df.sales.values + intercept
lines
```

```python
df.category.unique()
```

```python
dot1 = go.Scatter(
    x=df[df.category == "Furniture"].sales,
    y=df[df.category == "Furniture"].profit,
    mode="markers",
    name = "Furniture"
)
dot2 = go.Scatter(
    x=df[df.category == "Office Supplies"].sales,
    y=df[df.category == "Office Supplies"].profit,
    mode="markers",
    name = "Office Supplies"
)
dot3 = go.Scatter(
    x=df[df.category == "Technology"].sales,
    y=df[df.category == "Technology"].profit,
    mode="markers",
    name = "Technology"
)

line = go.Scatter(
    x=df.sales,
    y=lines,
    name = "Linear Trend Line"
)

data = [dot1, dot2, dot3, line]
layout={
    "title":{
        "text":"Sales VS Profit",
        "x" : 0.5
    },
    
    "xaxis" : {
        "title" : "Sales",
         "range": [0, 7500],
    },
    
    "yaxis" : {
        "title" : "Profit",
        "range": [-2500, 3000],
    },
   
}

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
